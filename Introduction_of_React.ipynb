Question 1 : What is React?

React is a popular open-source JavaScript library developed by Meta (Facebook) that is used to build fast, interactive, and reusable user interfaces for web and mobile applications. It focuses on creating components that update efficiently when data changes, making it ideal for modern, dynamic apps.

Key Features of React

Component-based:- Applications are built from small, reusable pieces of UI called components.

JSX syntax:- A JavaScript extension that allows developers to write HTML-like code directly in JavaScript.

Virtual DOM:- React updates only the parts of the UI that change, improving performance.

Unidirectional data flow:- Data moves in one direction, making apps easier to debug and maintain.

Question 3 : What is JSX and why is it useful in React?


JSX (JavaScript XML) is a syntax extension for JavaScript that allows developers to write HTML-like markup directly inside JavaScript code. It is widely used in React because it makes UI components more intuitive, readable, and easier to maintain.

Example :

```python
const element = <h1>Hello, world!</h1>;
```

This JSX code is transformed into:

```python
const element = React.createElement('h1', null, 'Hello, world!');
```

Why JSX is Useful in React:-

Readable UI code: Developers can visualize the UI structure directly in the code.

Dynamic rendering: Curly braces {} allow embedding JavaScript expressions inside markup.

Component integration: Keeps markup and rendering logic together, ensuring consistency.

Error handling: JSX enforces stricter rules than HTML (e.g., closing tags), reducing bugs.

Tooling support: Modern build tools like Babel and Webpack automatically compile JSX into JavaScript.

Question 4 : Explain how props support React’s one-way data flow.


Props in React enforce one-way data flow by allowing parent components to pass data down to child components, while preventing children from directly modifying that data. This ensures predictable, maintainable, and debuggable applications.

How Props Support One-Way Data Flow :


Parent-to-child communication: Data flows downward from parent components to child components via props.

Read-only nature: Props cannot be changed by child components, preserving data integrity.

Controlled updates: Only the parent component can update its state, which then re-renders children with new props.

Predictability: Since data only flows in one direction, it’s easier to trace and debug issues.

Example: One-Way Data Flow with Props :

```python
function Parent() {
  const [user, setUser] = React.useState("Alice");
  return <Child name={user} />;
}

function Child({ name }) {
  return <h1>Hello, {name}!</h1>;
}
```

The Parent owns the state (user).

The Child receives name as a prop but cannot modify it.

If the parent updates user, React re-renders the child with the new value.

Question 5 : Create a functional component ClickMeButton that renders a button. It
accepts onClick and label props. When clicked, it executes the passed onClick
handler.

```python
import React from 'react';

function ClickMeButton({ onClick, label }) {
  return (
    <button onClick={onClick}>
      {label}
    </button>
  );
}

export default ClickMeButton;
```

Explanation :


Functional component: Defined as a simple function returning JSX.

Props: Accepts two props:

onClick: A function passed from the parent, executed when the button is clicked.

label: A string displayed inside the button.

Event handling: The onClick prop is attached to the button’s onClick event.

```python
import React from 'react';
import ReactDOM from 'react-dom';
import ClickMeButton from './ClickMeButton';

function App() {
  const handleClick = () => {
    alert("Button was clicked!");
  };

  return (
    <div>
      <ClickMeButton onClick={handleClick} label="Click Me!" />
    </div>
  );
}

ReactDOM.render(<App />, document.getElementById('root'));
```

Question 6 : Given const fruits = ['Apple', 'Banana', 'Cherry'], write a
component FruitList that renders an unordered list of these fruits using .map() and
includes a key.

```python
import React from 'react';

const fruits = ['Apple', 'Banana', 'Cherry'];

function FruitList() {
  return (
    <ul>
      {fruits.map((fruit, index) => (
        <li key={index}>{fruit}</li>
      ))}
    </ul>
  );
}

export default FruitList;
```

Explanation:

Array mapping: The .map() method iterates over the fruits array and returns a li element for each fruit.

Key prop: Each list item needs a unique key to help React efficiently update and re-render lists. Here, the index is used as the key.

Unordered list: The ul element wraps the list items.

```python
import React from 'react';
import ReactDOM from 'react-dom';
import FruitList from './FruitList';

ReactDOM.render(<FruitList />, document.getElementById('root'));
```

This will render:

Apple

Banana

Cherry

Question 8 : Write a component UserList that receives a users prop—an array of
objects each with id and name. Render each user’s name in a p tag, using map()
and key.


```python
import React from 'react';

function UserList({ users }) {
  return (
    <div>
      {users.map(user => (
        <p key={user.id}>{user.name}</p>
      ))}
    </div>
  );
}

export default UserList;
```

 Explanation :


Props: The users prop is passed from the parent component.

Array mapping: .map() iterates over the array and returns a p element for each user.

Key attribute: Each p uses user.id as a unique key, ensuring React can efficiently track updates.

Reusable component: This design allows you to pass any array of user objects and render them consistently.

```python
import React from 'react';
import ReactDOM from 'react-dom';
import UserList from './UserList';

const users = [
  { id: 1, name: "Alice" },
  { id: 2, name: "Bob" },
  { id: 3, name: "Charlie" }
];

ReactDOM.render(<UserList users={users} />, document.getElementById('root'));
```

This will render:

Alice

Bob

Charlie

Question 9 : Create an App component that holds an array of numbers, passes them to
a NumberList child component for rendering, and passes a click handler function to
log the clicked number. Show one-way data flow and event handling.

```python
import React from 'react';

// Child component
function NumberList({ numbers, onNumberClick }) {
  return (
    <ul>
      {numbers.map(num => (
        <li key={num}>
          <button onClick={() => onNumberClick(num)}>
            {num}
          </button>
        </li>
      ))}
    </ul>
  );
}

// Parent component
function App() {
  const numbers = [1, 2, 3, 4, 5];

  const handleNumberClick = (num) => {
    console.log("Clicked number:", num);
  };

  return (
    <div>
      <h1>Number List</h1>
      <NumberList numbers={numbers} onNumberClick={handleNumberClick} />
    </div>
  );
}

export default App;
```

Explanation :

App component: Holds the array of numbers and defines the click handler.

NumberList child: Receives numbers and onNumberClick as props.

One-way data flow: Data (numbers) flows from parent → child. The child cannot modify the parent’s data directly.

Event handling: The child calls the parent’s handleNumberClick function when a button is clicked, passing the clicked number back up.

Question 10 : Design a component ProductList that receives products (array of
objects with id, name, price). Render a list of product names and prices using
.map(). Each item when clicked should call a prop onProductSelect(product).
Include the proper key.


```python
import React from 'react';

function ProductList({ products, onProductSelect }) {
  return (
    <ul>
      {products.map(product => (
        <li key={product.id}>
          <button onClick={() => onProductSelect(product)}>
            {product.name} - ${product.price}
          </button>
        </li>
      ))}
    </ul>
  );
}

export default ProductList;
```

Explanation :

Props:

products: An array of objects { id, name, price }.

onProductSelect: A callback function passed from the parent.

Array mapping: .map() iterates over the products and renders each as a list item.

Event handling: Clicking a product calls onProductSelect(product), sending the selected product back to the parent.

```python
import React from 'react';
import ReactDOM from 'react-dom';
import ProductList from './ProductList';

function App() {
  const products = [
    { id: 1, name: "Laptop", price: 1200 },
    { id: 2, name: "Phone", price: 800 },
    { id: 3, name: "Headphones", price: 150 }
  ];

  const handleProductSelect = (product) => {
    console.log("Selected product:", product);
  };

  return (
    <div>
      <h1>Available Products</h1>
      <ProductList products={products} onProductSelect={handleProductSelect} />
    </div>
  );
}

ReactDOM.render(<App />, document.getElementById('root'));
```

One-way data flow: Parent (App) owns the product data and passes it down.

Child communication: Child (ProductList) notifies the parent of user actions via the callback.

Predictability: The parent controls state and logic, while the child focuses on rendering.
